```python
""" Veri madenciliği teknikleri, veriden anlamlı bilgileri ortaya çıkarmak için farklı yaklaşımlar sunar. Baltık Denizi'ndeki kaza verilerini analiz ederken, bu projede kullanabileceğimiz bazı veri madenciliği teknikleri şunlardır:

1. Sınıflandırma (Classification)
Kullanım Alanı: Kazaların türlerine göre sınıflandırılması, riskli kaza türlerinin belirlenmesi için kullanılabilir.
Nasıl Uygulanabilir: Makine öğrenmesi algoritmalarını (örneğin, karar ağaçları veya lojistik regresyon) kullanarak kazaları risk düzeylerine göre sınıflandırabiliriz. Böylece, belirli bir ülkede veya dönemde hangi kaza türlerinin daha riskli olduğu hakkında çıkarım yapabiliriz.
2. Kümeleme (Clustering)
Kullanım Alanı: Benzer kaza türleri veya kazaların ortak özellikleri bazında kümelenmesi sağlanabilir.
Nasıl Uygulanabilir: Kaza verilerini konum, yıl ve kaza türlerine göre kümelendirerek, benzer özelliklere sahip kazaların hangi bölgelerde yoğunlaştığını belirleyebiliriz. K-ortalamalar (K-means) gibi algoritmalar bu amaç için uygundur.
3. Regresyon Analizi (Regression Analysis)
Kullanım Alanı: Belirli bir değişkenin (örneğin kaza oranı) başka faktörlerle ilişkisini anlamak için kullanılır.
Nasıl Uygulanabilir: Yıl veya deniz trafiği yoğunluğu gibi değişkenlerin kaza oranları üzerindeki etkisini inceleyebiliriz. Örneğin, yıllara göre kazaların azalıp azalmadığını veya belirli dönemlerde artış gösterip göstermediğini değerlendirebiliriz.
4. Zaman Serisi Analizi (Time Series Analysis)
Kullanım Alanı: Zaman içinde kaza oranlarının nasıl değiştiğini anlamak için kullanılır.
Nasıl Uygulanabilir: Kazaların yıl bazında analiz edilerek artış ve azalış eğilimlerinin ortaya konması, kaza oranlarını etkileyen mevsimsel veya yıllık faktörlerin belirlenmesi sağlanabilir. Örneğin, yıllık kaza oranlarındaki değişiklikler incelenerek belirli dönemlerde artış olup olmadığını analiz edebiliriz.
5. Birliktelik Analizi (Association Rule Mining)
Kullanım Alanı: Farklı kaza türleri arasındaki ilişkileri ve hangi kazaların birlikte gerçekleşme eğiliminde olduğunu anlamak için kullanılır.
Nasıl Uygulanabilir: Kazaların sebepleri ve sonuçları arasındaki ilişkileri analiz ederek, bazı kaza türlerinin birlikte mi gerçekleştiğini veya belirli bir kaza türünün başka bir kaza türü ile bağlantılı olup olmadığını ortaya çıkarabiliriz.

"""




import pandas as pd


file_path = 'output_file.csv'
data = pd.read_csv(file_path)


data.head()
```

```python
"""Kümeleme analizi yaparak, benzer özelliklere sahip kazaların hangi gruplarda yoğunlaştığını inceleyebiliriz. Baltık Denizi'ndeki kazaları kazanın türü, yılı ve ülkesi gibi özellikler temelinde kümelendirerek, ortak özelliklere sahip kaza kümeleri ortaya çıkarabiliriz.

Bu analizi yapmak için genellikle K-Means veya DBSCAN gibi algoritmalar kullanılır. Burada:

Veri Hazırlığı: Kaza türü, yıl, ülke gibi niteliklerden oluşan veriyi sayısal hale getirip kümeleme için uygun hale getireceğiz.
Kümeleme Uygulaması: Uygun bir kümeleme algoritması (K-Means) ile kazaların kümelere ayrılmasını sağlayacağız.
Sonuçların Görselleştirilmesi: Ortaya çıkan kümeleri görselleştirip, hangi özelliklerdeki kazaların aynı kümelerde yer aldığını gözlemleyeceğiz."""

from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt

# Prepare the data for clustering by selecting relevant columns and encoding categorical data
clustering_data = filtered_data[['Year', 'Country', 'Acc_Type']].dropna()

# Encode categorical columns 'Country' and 'Acc_Type'
label_enc_country = LabelEncoder()
label_enc_acc_type = LabelEncoder()
clustering_data['Country_encoded'] = label_enc_country.fit_transform(clustering_data['Country'])
clustering_data['Acc_Type_encoded'] = label_enc_acc_type.fit_transform(clustering_data['Acc_Type'])

# Select only numeric columns for clustering
clustering_data_numeric = clustering_data[['Year', 'Country_encoded', 'Acc_Type_encoded']]

# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data_numeric)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=0)  # Number of clusters can be adjusted
clustering_data['Cluster'] = kmeans.fit_predict(scaled_data)

# Cluster names based on identified trends (you can adjust based on findings)
cluster_names = {
    0: "High Collision Rate (2000s, Sweden)",
    1: "High Contact Rate (2010s, Germany)",
    2: "Mixed Accidents (Various countries)",
    3: "Rising Machinery Damage (2018+, Finland)"
}

# Visualize the clustering result (Year vs. Country with clusters)
plt.figure(figsize=(14, 8))
scatter = plt.scatter(clustering_data['Year'], clustering_data['Country_encoded'], c=clustering_data['Cluster'], cmap='viridis')
plt.colorbar(label='Cluster')
plt.xlabel('Year')
plt.ylabel('Country (Encoded)')
plt.title('Clustering of Accidents by Year and Country')

# Annotate clusters with descriptive names
for cluster_id, cluster_name in cluster_names.items():
    plt.text(1985, cluster_id * 5, f"{cluster_id}: {cluster_name}", fontsize=10, bbox=dict(facecolor='white', alpha=0.5), ha='left')

plt.tight_layout()
plt.show()


"""
Yüksek Çarpışma Oranı (High Collision Rate): İsveç ve bazı çevre ülkelerde, özellikle 2000'li yıllarda çarpışma kazalarının yoğunlaştığı bir küme göze çarpıyor. Bu, o dönemde deniz trafiği veya güvenlik önlemleriyle ilgili iyileştirilmesi gereken alanlar olduğunu gösterebilir.
Yüksek İrtibat Oranı (High Contact Rate): Almanya başta olmak üzere, 2010'lu yıllarda başka bir kaza türü olan "contact" kazalarının yoğunlaştığı bir küme mevcut. Bu kazaların sık yaşandığı dönem ve ülkelerde deniz güvenliği tedbirlerinin gözden geçirilmesi gerekebilir.
Karma Kazalar (Mixed Accidents): Birden fazla ülkeyi ve kaza türünü içeren bir küme daha var. Bu kümede farklı türde kazaların görüldüğü çeşitli ülkeler yer alıyor. Bu küme, belirli bir kaza türünden ziyade genel bir kaza eğilimi gösteriyor olabilir.
Yükselen Makinada Arıza (Rising Machinery Damage): 2018 yılından sonra Finlandiya ve yakın ülkelerde makinada arızaların artış gösterdiği bir küme gözlemleniyor. Bu arızalar, teknoloji eskimesi veya bakım eksiklikleri gibi sorunlardan kaynaklanıyor olabilir.

"""
```

```python
""" regresyon analizi ile kaza oranlarının zamana veya belirli değişkenlere göre nasıl değiştiğini inceleyebiliriz. Bu proje için örneğin, yıllara göre kaza oranlarındaki değişiklikleri inceleyerek bir eğilim olup olmadığını analiz edebiliriz."""
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

# Prepare data: Group accidents by year to get yearly accident counts
yearly_accidents = filtered_data.groupby('Year').size().reset_index(name='Accident_Count')

# Reshape the data for linear regression model
X = yearly_accidents['Year'].values.reshape(-1, 1)  # Independent variable (Year)
y = yearly_accidents['Accident_Count'].values       # Dependent variable (Accident count)

# Initialize and fit the linear regression model
linear_reg = LinearRegression()
linear_reg.fit(X, y)

# Predictions
y_pred = linear_reg.predict(X)

# Model evaluation metrics
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

# Plotting the results
plt.figure(figsize=(14, 8))
plt.scatter(X, y, color='blue', label='Actual Accident Counts')
plt.plot(X, y_pred, color='red', linestyle='--', label='Linear Regression Fit')
plt.title('Yearly Accident Trends with Linear Regression')
plt.xlabel('Year')
plt.ylabel('Accident Count')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Display model performance
r2, mse

```
